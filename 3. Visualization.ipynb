```python
# load dataset into Pandas DataFrame
file = pd.ExcelFile('Iris.xlsx')
df = pd.read_excel(file, names=['sepal length','sepal width','petal length','petal width','target'])
df
```

```python
df = df.drop('sepal length', axis =1)
#df = df.drop('sepal width', axis =1)
#df = df.drop('petal length', axis =1)
#df = df.drop('target', axis =1)
df = df.rename(columns = {"petal width": "vnfjdm"})
df
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('sepal length', fontsize = 15)
ax.set_ylabel('sepal width', fontsize = 15)
ax.set_title('Data visualization', fontsize = 20)
targets = [0, 1, 2]
#targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = df['target'] == target
    ax.scatter(df.loc[indicesToKeep, 'sepal length'],
               df.loc[indicesToKeep, 'sepal width'],
               c = color,
               s = 50)
ax.legend(targets)
ax.grid()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize = (8,8))
ax = Axes3D(fig, elev=-130, azim=70) 
ax.set_xlabel('sepal length', fontsize = 15)
ax.set_ylabel('sepal width', fontsize = 15)
ax.set_zlabel('petal length', fontsize = 15)
ax.set_title('Data visualization 3D', fontsize = 20)
targets = [0, 1, 2]
#targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = df['target'] == target
    ax.scatter(df.loc[indicesToKeep, 'sepal length'],
               df.loc[indicesToKeep, 'sepal width'],
               df.loc[indicesToKeep, 'petal length'],
               edgecolor='k',
               alpha = 1,
               c = color,
               s = 50)
ax.legend(targets)
ax.grid()
```

```python
import numpy as np
import pandas as pd

%matplotlib inline

# load dataset into Pandas DataFrame
file = pd.ExcelFile('Iris.xlsx')
df = pd.read_excel(file, names=['sepal length','sepal width','petal length','petal width','target'])

features = ['sepal length', 'sepal width', 'petal length', 'petal width']

# Separating out the features
x = df.loc[:, features].values
#print(x)

# Separating out the target
y = df.loc[:,['target']].values
#print(y)
```

```python
from sklearn.manifold import TSNE
tsne = TSNE (n_components = 2, perplexity = 30, n_iter=1000, random_state = 0, learning_rate = 100)
x_2D = tsne.fit_transform(x)
```

```python
tsneDf = pd.DataFrame(data = x_2D, columns = ['dim 1', 'dim 2'])
print(tsneDf)
```

```python
final_tsneDf = pd.concat([tsneDf, df[['target']]], axis = 1)
final_tsneDf
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('dim 1', fontsize = 15)
ax.set_ylabel('dim 2', fontsize = 15)
ax.set_title('2D TSNE', fontsize = 20)
targets = [0, 1, 2]
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = final_tsneDf['target'] == target
    ax.scatter(final_tsneDf.loc[indicesToKeep, 'dim 1']
               , final_tsneDf.loc[indicesToKeep, 'dim 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE (n_components = 3, perplexity = 50, n_iter=5000, random_state = 0, learning_rate = 250)
x_3D = tsne.fit_transform(x)
```

```python
tsneDf = pd.DataFrame(data = x_3D, columns = ['dim 1', 'dim 2', 'dim 3'])
tsneDf
```

```python
final_tsneDf = pd.concat([tsneDf, df[['target']]], axis = 1)
final_tsneDf
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize = (8,8))
ax = Axes3D(fig, elev=-160, azim=70) 
ax.set_xlabel('dim 1', fontsize = 15)
ax.set_ylabel('dim 2', fontsize = 15)
ax.set_zlabel('dim 3', fontsize = 15)
ax.set_title('t-SNE', fontsize = 20)
targets = [0, 1, 2]
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = final_tsneDf['target'] == target
    ax.scatter(final_tsneDf.loc[indicesToKeep, 'dim 1'],
               final_tsneDf.loc[indicesToKeep, 'dim 2'],
               final_tsneDf.loc[indicesToKeep, 'dim 3'],
               c = color, alpha = 1,
               s = 50)
ax.legend(targets)
ax.grid()
```
