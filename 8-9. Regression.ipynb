```python
# generate random data-set
np.random.seed(0)
x = np.random.rand(100, 1)
y = 2 + 3 * x + np.random.rand(100, 1)
```

```python
from sklearn.linear_model import LinearRegression

# инициализация модели
regression_model = LinearRegression()
# Обучение модели
regression_model.fit(x, y)
# результат работы модели
y_predicted = regression_model.predict(x)


# printing values
print('Coef:' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)
```

```python
# Рисуем рисунок
plt.scatter(x, y, s=10)
plt.xlabel('x')
plt.ylabel('y')

# predicted values
plt.plot(x, y_predicted, color='r')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# оценка модели
mae = mean_absolute_error(y, y_predicted)
rmse = mean_squared_error(y, y_predicted)
r2 = r2_score(y, y_predicted)

print('Mean Absolute Error:', mae)
print('Root mean squared error: ', rmse)
print('R2 score: ', r2)

```

## Сигмоїдальна функція

Припустимо, ми хотіли класифікувати наші дані на дві категорії: негативні і позитивні. На відміну від лінійної регресії, де ми хочемо передбачити безперервне значення, ми хочемо, щоб наш класифікатор передбачав ймовірність того, що дані позитивні (1) або негативні (0). Для цього ми будемо використовувати функцію Sigmoid:

```python
from IPython.display import Image
Image("f2.png")
```

Створимо для цього функцію:

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

Якщо ми побудуємо функцію, ми зауважимо, що в міру наближення входу до нескінченності вихід наближається до 1, а коли вхід наближається до мінус нескінченності - вихід наближається до 0.

```python
x = np.linspace(-10, 10, 200)
plt.plot(x, sigmoid(x))
plt.axvline(x=0, color='k', linestyle='--');
plt.title("Sigmoid");
```

Передавши добуток наших вхідних даних і параметрів в сигмоїдальну функцію g, ми можемо сформувати прогноз h ймовірності того, що вхід х буде класифікований позитивно.

## Функція вартості
Коли ми проводили лінійну регресію, ми використовували середньоквадратичну помилку в якості нашої функції вартості. Це добре працює для регресії, але для класифікації ми хочемо використовувати функцію Cross Entropy Loss або функцію перехресної втрати ентропії J

```python
from IPython.display import Image
Image("f1.png")
```

Ми можемо зрозуміти, як працює функція Cross Entropy Loss, за графіком.
Наша класифікація y може бути 1 або 0:

```python
h = np.linspace(0, 1)[1:-1]
for y in [0, 1]:
    plt.plot(h, -y * np.log(h) - (1 - y) * np.log(1 - h), label=f"y={y}")
plt.title("Cross Entropy Loss")
plt.xlabel('$h_ {\\theta}(x)$'); 
plt.ylabel('$J(\\theta)$');
plt.legend();
```

Ми можемо бачити ситуацію, в якій прогноз, що відповідає вірній класифікації, матиме вартість 0 (помаранчевий графік) і наближається до нескінченності, коли прогноз наближається до неправильної класифікації (синій графік).

## Функція градієнта

Візьмемо похідну функції вартості за нашими параметрами як функції градієнта для нашого градієнтного спуску:

```python
Image("f4.png")
```

Тепер ми можемо написати одну функцію Python, що повертає як нашу вартість, так і градієнт:

```python
#gradient function
def cost(theta, x, y):
    h = sigmoid(x @ theta)
    m = len(y)
    cost = 1 / m * np.sum(-y * np.log(h) - (1 - y) * np.log(1 - h))
    grad = 1 / m * ((y - h) @ x)
    return cost, grad
```

Так само, як лінійна регресія з градієнтним спуском, ми инициализируем наші параметри θ,
спочатку задаємо їх нульовими, потім оновлюємо їх на кожній епосі, використовуючи:

```python
Image("f5.png")
```

Ще одне міркування: метод класифікації працює тільки з двома мітками класу: позитивним і негативним. Щоб класифікувати більше двох міток, ми будемо використовувати так звану стратегію «один проти інших»: для кожної мітки класу ми підженемо набір параметрів, де ця мітка класу позитивна, а решта негативні. Потім ми можемо сформувати прогноз, вибравши максимальну гіпотезу hθ (x) для кожного набору параметрів.

```python
def fit(x, y, max_iter=5000, alpha=0.1):
    x = np.insert(x, 0, 1, axis=1)
    thetas = []
    classes = np.unique(y)
    costs = np.zeros(max_iter)

    for c in classes:
        # классификация один против всех
        binary_y = np.where(y == c, 1, 0)

        theta = np.zeros(x.shape[1])
        for epoch in range(max_iter):
            costs[epoch], grad = cost(theta, x, binary_y)
            theta += alpha * grad

        thetas.append(theta)
    return thetas, classes, costs
```

Ми також можемо написати функцію передбачення, яка передбачає мітку класу, використовуючи максимальну гіпотезу hθ(x)

```python
def predict(classes, thetas, x):
    x = np.insert(x, 0, 1, axis=1)
    preds = [np.argmax(
    [sigmoid(xi @ theta) for theta in thetas]
    ) for xi in x]
    return [classes[p] for p in preds]
```

Перевіримо роботу на ірисах

```python
# Импортируем библиотеки
from sklearn.datasets import load_iris

# Загружаем датасет
iris = load_iris()
x = np.array(iris.data)
y = np.array(iris.target)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, 
                                                    random_state = 7)
```

```python
plt.scatter(x_train[:,2], x_train[:, 3], c=y_train, alpha=0.5)
plt.xlabel("Petal Length (cm)"); plt.ylabel("Petal Width (cm)");
```

Модель підвищує свою вартість в кожну епоху.

```python
thetas, classes, costs = fit(x_train[:, 2:], y_train)
plt.plot(costs)
plt.xlabel('Number Epochs'); plt.ylabel('Cost');
```

```python
plt.scatter(x_train[:,2], x_train[:, 3], c=y_train, alpha=0.5)
plt.xlabel("Petal Length (cm)"); plt.ylabel("Petal Width (cm)")
for theta in [thetas[0],thetas[2]]:
    j = np.array([x_train[:, 2].min(), x_train[:, 2].max()])
    k = -(j * theta[1] + theta[0]) / theta[2]
    plt.plot(j, k, color='k', linestyle="--")
```

```python
def score(classes, theta, x, y):
    return (predict(classes, theta, x) == y).mean()

print(f"Train Accuracy: {score(classes, thetas, x_train[:, 2:], y_train):.3f}")
print(f"Test Accuracy: {score(classes, thetas, x_test[:, 2:], y_test):.3f}")
```

```python
thetas, classes, costs = fit(x_train, y_train)
print(f"Train Accuracy: {score(classes, thetas, x_train, y_train):.3f}")
print(f"Test Accuracy: {score(classes, thetas, x_test, y_test):.3f}")
```
